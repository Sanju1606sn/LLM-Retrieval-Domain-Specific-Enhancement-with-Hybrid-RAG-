## 3. **add some helper functions**

````python
def extract_json_from_response(text: str) -> dict:
    """
    Parses a variety of LLM outputs into a valid Python dictionary.
    Args:
        text (str): LLM output string that should represent a JSON/dict object.
    Returns:
        dict: Parsed Python dictionary.
    Raises:
        ValueError: If the input cannot be parsed into a dictionary.
    """
    # Trim and clean up common LLM issues
    text = text.strip()

    # Remove code block markers if present
    text = re.sub(r"^```(?:json)?\s*|\s*```$", "", text, flags=re.IGNORECASE).strip()

    # Try JSON decoding first
    try:
        return json.loads(text.replace("'", '"'))
    except json.JSONDecodeError:
        msg = "JSON Decode Error"

    # Try Python literal_eval
    try:
        result = ast.literal_eval(text)
        if isinstance(result, dict):
            return result
    except (SyntaxError, ValueError):
        msg = "Syntax Error, ast error"

    # Try extracting dictionary-like structure with regex
    match = re.search(r"\{.*\}", text, re.DOTALL)
    if match:
        try:
            return ast.literal_eval(match.group())
        except Exception:
            msg = "Regex Error"

    raise ValueError("Unable to parse LLM output into JSON in agent due " + msg)

def get_weather(city: str, date_str: str) -> Union[Dict[str, Union[str, float]], str]:
    """
    Retrieves the weather forecast for a specific city and date using 7Timer API 
    and geolocation from OpenStreetMap.

    This function performs three main tasks:
    1. Converts the input date to a datetime object and validates it.
    2. Uses the OpenStreetMap Nominatim API to get geographic coordinates for the city.
    3. Queries the 7Timer weather API to get forecast data and extracts weather 
       details for the specified date.

    Args:
        city (str): Name of the city to get the weather forecast for (e.g., "Paris").
        date_str (str): Date as a string in "YYYY-MM-DD" format or a `datetime` object.

    Returns:
        Union[Dict[str, Union[str, float]], str]:
            - On success: A dictionary containing:
                - 'city' (str): The input city.
                - 'date' (str): The validated and formatted target date.
                - 'min_temp' (float): Minimum forecast temperature (Â°C).
                - 'max_temp' (float): Maximum forecast temperature (Â°C).
                - 'weather_summary' (str): Most frequent weather description for the day.
            - On failure: A string describing the error encountered (e.g., invalid date, 
              no forecast data, or network issues).

    Example:
        >>> get_weather("Barcelona", "2025-04-25")
        {
            'city': 'Barcelona',
            'date': '2025-04-25',
            'min_temp': 14.0,
            'max_temp': 21.0,
            'weather_summary': 'clear'
        }
    """
    try:
        if isinstance(date_str, datetime):
            target_date = date_str.date()
        elif isinstance(date_str, str):
            target_date = datetime.strptime(date_str, "%Y-%m-%d").date()
        else:
            return "Error: Provided date must be a string or datetime object."

        if target_date < datetime.now().date():
            return "Error: Date cannot be in the past."

    except Exception as e:
        return f"Error parsing date: {e}"


    # Get coordinates using OpenStreetMap
    try:
        encoded_city = quote(city[0])
        geo_url = f"https://nominatim.openstreetmap.org/search?q={encoded_city}&format=jsonv2&limit=1"
        headers = {'User-Agent': 'WeatherForecastApp/1.0'}
        geo_res = requests.get(geo_url, headers=headers)
        geo_res.raise_for_status()
        geo_data = geo_res.json()
        if not geo_data:
            return f"Error: Could not find coordinates for '{city}'"
        lat = float(geo_data[0]['lat'])
        lon = float(geo_data[0]['lon'])
    except Exception as e:
        return f"Error getting coordinates: {e}"

    # Get weather data from 7timer API
    try:
        url = f"http://www.7timer.info/bin/api.pl?lon={lon}&lat={lat}&product=civil&output=json"
        weather_res = requests.get(url)
        weather_res.raise_for_status()
        weather_data = weather_res.json()
    except Exception as e:
        return f"Error fetching weather data: {e}"

    # Parse forecast data
    results = []
    for item in weather_data.get("dataseries", []):
        timepoint = item["timepoint"]
        base_date = datetime.strptime(str(weather_data["init"]), "%Y%m%d%H")
        forecast_time = base_date + timedelta(hours=timepoint)
        if forecast_time.date() == target_date:
            results.append({
                "temp": item["temp2m"],
                "weather_desc": item["weather"],
                "rh": int(item["rh2m"].strip('%')) if "rh2m" in item else None,
                "wind": item.get("wind10m", {}).get("speed", None)
            })

    if not results:
        return f"No forecast data available for {target_date} in {city}"

    min_temp = min(r["temp"] for r in results)
    max_temp = max(r["temp"] for r in results)
    weather_summary = _get_most_frequent_weather(results)

    return {
        "city": city,
        "date": str(target_date),
        "min_temp": min_temp,
        "max_temp": max_temp,
        "weather_summary": weather_summary
    }

def _get_most_frequent_weather(results):
    freq = {}
    for r in results:
        w = r.get("weather_desc", "Unknown")
        freq[w] = freq.get(w, 0) + 1

    return max(freq, key=freq.get) if freq else "Unknown"
````

# Agents

## **ğŸ¨ Accommodation Agent**  
An intelligent hotel recommendation engine that combines real-world travel data, vector search, and large language models to return **personalized, context-aware accommodation suggestions**.

---

### **1. Real-World Reviews with Kaggle Dataset**  
The agent starts by using the **515k Hotel Reviews in Europe** dataset from Kaggle, which includes hotel names, positive and negative reviews, and addresses. We use `pandas` to filter, aggregate, and rank hotels based on:\n
- Relevance to the user's preferences  
- Number of available reviews

> **Why it matters:** Recommendations are grounded in actual, crowd-sourced data â€” not generated from scratch.  
> **Note:** If the requested city isnâ€™t covered, we fall back to an LLM-based prompt that avoids generating fake hotel names while still providing viable suggestions.

---

### **2. Gemini Embeddings for Intent Matching**  
We define a custom embedding function using **Geminiâ€™s `text-embedding-004` model**, which creates semantic vector representations for both:\n
- User search preferences  
- Aggregated hotel reviews

> **Why it matters:** This enables true intent-based search, beyond simple keyword matching.

---

### **3. Fast Retrieval with ChromaDB**  
We use **ChromaDB** to manage our vector database. The system:\n
- Indexes hotel reviews and metadata as embeddings  
- Performs fast similarity search using cosine distance  
- Summarizes all top reviews for each hotel before embedding to reduce vector size

> **Why it matters:** The system is fast, scalable, and retrieves relevant results with minimal overhead.

---

### **4. Geolocation Resolution with GeoPy**  
To resolve potential ambiguities (e.g., multiple cities named â€œParisâ€), we use `GeoPy` to infer the country based on the city name. This allows us to more accurately filter hotel results within the correct region.

> **Why it matters:** Improves the quality and relevance of location-based filtering.

---

### **5. Retrieval-Augmented Generation (RAG)**  
After retrieving the most relevant hotel data, we feed that content into a Gemini prompt, asking it to generate a clear, friendly summary of the results.

> **Why it matters:** This hybrid approach blends data accuracy with LLM-generated readability.

---

### **6. Long-Context Gemini Prompting**  
Thanks to Geminiâ€™s long context support, we can include detailed summaries of multiple hotels in a single prompt.

> **Why it matters:** Enables smarter comparisons and more coherent recommendations across multiple listings.

---

### **In Summary**  
This agent intelligently combines structured data, semantic search, and generative AI to help users find the right place to stay â€” based on what truly matters to them.


````python


class GeminiEmbeddingFunction(EmbeddingFunction):
    # Specify whether to generate embeddings for documents, or queries
    document_mode = True

    def __init__(self):
        super().__init__()

    @retry.Retry(predicate=is_retriable)
    def __call__(self, input: Documents) -> Embeddings:
        if self.document_mode:
            embedding_task = "retrieval_document"
        else:
            embedding_task = "retrieval_query"

        response = client.models.embed_content(
            model="models/text-embedding-004",
            contents=input,
            config=types.EmbedContentConfig(
                task_type=embedding_task,
            ),
        )
        return [e.values for e in response.embeddings]
        
class AccommodationSearchAgent:
    """
    An intelligent hotel recommendation agent that combines real-world travel data, vector search,
    and large language models to return personalized accommodation suggestions.

    Key Features:
        - Loads a Kaggle dataset of 500K+ hotel reviews from European cities.
        - Filters reviews based on user-specified destination and inferred country using geolocation.
        - Aggregates and summarizes hotel reviews for each property.
        - Embeds hotel summaries using Google's Gemini embedding model.
        - Stores and queries embeddings via ChromaDB for fast similarity-based retrieval.
        - Formats and summarizes search results using Gemini's chat model for a friendly user experience.

    Args:
        csv_path (str): Path to the hotel review CSV dataset. Defaults to the downloaded Kaggle dataset.

    Methods:
        _df_reviews(): Loads and caches the hotel review dataset as a Pandas DataFrame.
        _get_country_from_city(city): Uses geopy to infer the country of a given city.
        _filter_country(df, query): Filters hotel reviews based on city and country match.
        _aggregate_hotels(df): Aggregates and ranks top hotels by review count.
        _index_top_hotels(top): Embeds and stores top hotels into the Chroma vector database.
        _similarity_search(query): Searches for hotels most relevant to user preferences using embeddings.
        _format_response(query, hits): Summarizes results using Gemini LLM in a user-friendly format.
        __call__(query): Entry point to generate recommendations based on user input.
    """
    def __init__(self, csv_path: str = path + "/Hotel_Reviews.csv"):
        self.csv_path = csv_path
        self.embed_fn = GeminiEmbeddingFunction()
        self.llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.1,google_api_key=GOOGLE_API_KEY )
        self.col = chromadb.Client(
                                    # Settings(
                                    # chroma_db_impl="duckdb+parquet",                   # use DuckDB backend
                                    # persist_directory="/kaggle/working/chroma_data",   # persistent path on Kaggle
                                    # anonymized_telemetry=False)                         # optional: suppress telemetry).get_or_create_collection("accommodation_reviews")
            
        ).get_or_create_collection("accommodation_reviews")
        self._df: Optional[pd.DataFrame] = None

    def _df_reviews(self) -> pd.DataFrame:
        if self._df is None:
            self._df = pd.read_csv(
                self.csv_path, engine="python", on_bad_lines="skip",
                encoding="utf-8", na_values=["NA"]
            )[["Hotel_Address", "Positive_Review", "Negative_Review", "Hotel_Name"]]
            self._df["CityCandidate"] = self._df["Hotel_Address"].str.split().apply(lambda x: " ".join(x[-3:]))
        return self._df.copy()
    
    def _get_country_from_city(self, city: str) -> str:
        geolocator = Nominatim(user_agent="city-country-resolver")
        location = geolocator.geocode(city)
        if location:
            return location.address.split(",")[-1].strip()
        return "Country not found"

    def _filter_country(self, df: pd.DataFrame, query: Dict) -> pd.DataFrame:
        city=query['destination']
        country= self._get_country_from_city(city)
        parsed = city+" "+country
        new_df = df[df["CityCandidate"].str.contains(parsed, case=False, na=False)]
        return new_df

    def _aggregate_hotels(self, df: pd.DataFrame) -> pd.DataFrame:
        grouped = (
            df.groupby("Hotel_Name", as_index=False)
            .agg(
                Hotel_Address=("Hotel_Address", "first"),
                Positive_Review=("Positive_Review", lambda x: " ".join(x.dropna().head(5))),
                Negative_Review=("Negative_Review", lambda x: " ".join(x.dropna().head(5))),
            )
        )
        grouped["review_count"] = df.groupby("Hotel_Name").size().values
        group=grouped.sort_values("review_count", ascending=False).head(10)
        return group

    def _index_top_hotels(self, top: pd.DataFrame):
        # use vectorized pandas ops
        docs, metas, ids = [], [], []
        top["doc"] = (
            top["Hotel_Name"] + " (" + top["Hotel_Address"] + "):\nPositive: " +
            top["Positive_Review"] + "\nNegative: " + top["Negative_Review"]
        )

        docs = top["doc"].tolist()
        metas = top[["Hotel_Name", "Hotel_Address", "review_count"]].rename(
            columns={"Hotel_Name": "hotel_name", "Hotel_Address": "hotel_address"}
        ).to_dict(orient="records")

        ids = top.index.astype(str).tolist()

        if not docs:
            return

        emb = self.embed_fn(docs)
        if not emb or len(emb) != len(docs):
            raise ValueError("Embedding failed or mismatch.")
        
        self.col.add(ids=ids, documents=docs, embeddings=emb, metadatas=metas)

    def _similarity_search(self, query: Dict, n: int = 3) -> List[dict]:
        destination = query.get("destination", "")
        preference = query.get("user_preference", "")
        full_query = f"Find hotels in {destination}. Preference: {preference}".strip()
        emb = self.embed_fn([full_query])
        res = self.col.query(query_embeddings=emb, n_results=n, include=["metadatas"])
        return res["metadatas"][0]


    def _format_response(self, query: Dict, hits: List[dict]) -> str:
        if not hits:
            return "No matching accommodations found for your query."
        text = f"User query: {query}\nHotel options:\n"
        for h in hits:
            text += f"- {h['hotel_name']} at {h['hotel_address']} (Reviews: {h['review_count']})\n"
        # Summarize with Gemini
        prompt = f"Please summarize the following hotel recommendations in a friendly tone:\n\n{text}\n\nSummary:"
        try:
            output= self.llm.invoke(prompt).content.strip()
            return output
        except Exception as e:
            print("Summarization fallback:", e)
            return text

    @traceable(name="Accomodation Recommender")
    def __call__(self, query: Dict) -> Dict[str, Any]:
        if not query:
            return {"response": "I need a search query."}
        df = self._filter_country(self._df_reviews(), query)
        if df.empty:
            prompt = f"""
        You are a helpful travel assistant. A user is looking for hotels in **{query["destination"]}**.

        They specifically care about: **{query["user preference"]}**

        Return a short, friendly list of 1 hotel in **{query["destination"]}** that best match the user's preference. 
        For each hotel, include:
        - Hotel name
        - Neighborhood or address
        - Why it matches the preference

        Do not invent fake hotel namesâ€”use real or plausible examples.
        Answer in Markdown format.
        """
            return {"response": self.llm.invoke(prompt).content.strip()}
        else:
            top = self._aggregate_hotels(df)
            self._index_top_hotels(top)
            hits = self._similarity_search(query)
            return {"response": self._format_response(query, hits)}
````

## **ğŸ—ºï¸ Activity Agent**  
Fetches real-time travel activities using **Google Programmable Search Engine (CSE)** and summarizes them with **Gemini** for personalized, destination-aware recommendations.

---

### **Why This Approach Works**

- Fits seamlessly into a **modular, multi-agent planning system**

### **In Summary**  
The Activity Agent combines real-time search, structured prompt engineering, and generative summarization to create **up-to-date, relevant, and engaging activity ideas** â€” all tailored to the travelerâ€™s destination.


````python
google_search_llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=GOOGLE_API_KEY)

def activity_search(agent_input: dict) -> str:
    """
    Uses Google Programmable Search Engine (CSE) to retrieve travel-related activity results,
    then summarizes them using Gemini LLM for user-friendly recommendations.

    Args:
        agent_input (dict): A dictionary that must include the key:
            - 'search_query' (str): A natural-language search string, e.g. "best museums in Paris".

    Returns:
        str: A summarized, friendly list of top activities based on real-time search results.
    """
    if not (GOOGLE_SEARCH_KEY and GOOGLE_CSE_ID):
        raise RuntimeError(
            "Set GOOGLE_SEARCH_API_KEY and GOOGLE_CSE_ID env vars first.")
    
    # Google Search API
    url = "https://www.googleapis.com/customsearch/v1"
    params = {
        "key": GOOGLE_SEARCH_KEY,
        "cx":  GOOGLE_CSE_ID,
        "q":   agent_input["search_query"],
        "num": 5,
    }

    try:
        raw = requests.get(url, params=params, timeout=10).json()
        items: List[Dict] = raw.get("items", [])
        # titles = [item.get("title") for item in items]  
    except Exception as e:
        traceback.print_exc()
        raise RuntimeError(f"Google Search error: {e}")

    if not items:
        return "I couldn't find any results for that search."

    # preparing prompt for LLM 
    compact = [{"title": it["title"], "link": it["link"], "snippet": it["snippet"]} for it in items]

    prompt = (
        f"You are a travel expert.\n"
        f"Pick the top 5 activities someone should do based on the "
        f"Google search results below and present them in a friendly style. "
        "Give each activity a oneâ€‘sentence description.\n\n"
        f"{json.dumps(compact, ensure_ascii=False, indent=2)}"
    )
    
    # summerization with LLM 
    summary = google_search_llm.invoke(prompt).content
    # Return as a dictionary with a tag (or using your message format) so that the router knows it is from tool.
    return summary
````

````python
class SillyTravelBriefing(BaseModel):
    """Structured output for the silly travel briefing."""
    destination: str 
    weather_summary: str 
    clothing_tip: str 
    cultural_tip: str 
    language_tip: str 
    safety_tip: str 
    activity_suggestion: str 
    closing_remark: str 

class TravelPreferences(BaseModel):
    """Input model for travel details."""
    destination: str 
    travel_date: str 
    duration: str 
    preferences: str 
    budget: Optional[str] 

google_llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=GOOGLE_API_KEY)
Google_Search = GoogleSearchAPIWrapper(google_api_key=GOOGLE_API_KEY,google_cse_id= GOOGLE_CSE_ID)

def parse_duration(duration_str: str) -> int:
    """Parses duration strings like '5 days', '1 week' into number of days."""
    duration_str = duration_str.lower().strip()
    days = 1 # Default to 1 day if parsing fails

    # Regex to find numbers followed by day/week/month variations
    match = re.match(r"(\d+)\s*(day|week|month)s?", duration_str)
    if match:
        num = int(match.group(1))
        unit = match.group(2)
        if unit == "day":
            days = num
        elif unit == "week":
            days = num * 7
        elif unit == "month":
            # Approximate month duration, adjust if needed
            days = num * 30
    elif "weekend" in duration_str:
        days = 2 # Or 3, depending on definition
    else:
        # Try converting directly if it's just a number
        days = int(duration_str)

    # Ensure minimum of 1 day and reasonable maximum (e.g., 16 for many free forecast APIs)
    days = max(1, min(days, 16))
    return days

def get_daily_forecast(destination: str, start_date: datetime, num_days: int) -> Optional[List[dict]]:
    """
    Fetches daily forecast data (min/max temp, description) for `num_days`.
    Replace `get_weather()` with your actual API logic.
    """
    try:
        forecast_data = []
        for i in range(num_days):
            current_date = (start_date + timedelta(days=i)).strftime('%Y-%m-%d')
            weather_data = get_weather(destination, current_date)
            
            if isinstance(weather_data, dict) and all(k in weather_data for k in ["date", "min_temp", "max_temp", "weather_summary"]):
                forecast_data.append({
                    'date': weather_data["date"],
                    'min_temp': weather_data["min_temp"],
                    'max_temp': weather_data["max_temp"],
                    'description': weather_data["weather_summary"]
                })
            else:
                print(f"âš ï¸ Skipping {current_date}: Incomplete or invalid data â†’ {weather_data}")
                continue

        return forecast_data
    
    except Exception as e:
        print(f"Error in get_daily_forecast: {e}")
        return None


def get_clothing_tip_for_range(min_temp, max_temp, description: str) -> str:
    """Generates clothing advice based on min/max temperature range and description."""
    tip = ""
    has_min = isinstance(min_temp, (int, float))
    has_max = isinstance(max_temp, (int, float))
    has_temps = has_min and has_max
    rain_possible = "rain" in description.lower() or "shower" in description.lower()

    if not has_temps:
        tip = "Temperature data unavailable. Pack layers and hope for the best!"
    elif max_temp < 5:
        tip = f"Very Cold (Low {min_temp}Â°C, High {max_temp}Â°C)! Bundle up: Heavy coat, thermal layers, hat, gloves, scarf are essential."
    elif max_temp < 12:
        tip = f"Cold (Low {min_temp}Â°C, High {max_temp}Â°C). Wear layers: Warm jacket or coat, sweater/fleece, long pants. Consider hat/gloves, especially if windy or near {min_temp}Â°C."
    elif max_temp < 18:
        tip = f"Cool (Low {min_temp}Â°C, High {max_temp}Â°C). A medium jacket or layered sweaters recommended. Long pants ideal. Layers help adjust from cooler {min_temp}Â°C periods."
    elif max_temp < 25:
        tip = f"Mild/Warm (Low {min_temp}Â°C, High {max_temp}Â°C). Variable! T-shirt with a light jacket/sweater for cooler {min_temp}Â°C times (mornings/evenings). Pants or shorts depending on the {max_temp}Â°C."
    elif max_temp < 30:
        tip = f"Warm (Low {min_temp}Â°C, High {max_temp}Â°C). T-shirts, light pants or shorts. A very light layer might be nice for evenings near {min_temp}Â°C."
    else: # max_temp >= 30
        tip = f"Hot (Low {min_temp}Â°C, High {max_temp}Â°C)! Pack light, breathable clothing (cotton, linen), shorts, tank tops. Stay hydrated and use sun protection."

    if has_temps and abs(max_temp - min_temp) >= 10:
         tip += " Significant temperature swing: Layers are key!"

    if rain_possible:
        tip += " Pack a raincoat or umbrella!"

    return tip
def get_multi_day_weather_and_clothing_context(destination: str, start_date_str: str, duration_str: str) -> str:
    """
    Fetches multi-day forecast and generates weather summary and clothing tips
    for the duration of the stay.
    """
    # Parse Start Date
    if isinstance(start_date_str, datetime):
        start_date_obj = start_date_str
    else:
        start_date_obj = date_parser.parse(start_date_str)
    
    # Parse Duration
    num_days = parse_duration(duration_str)

    # Get Forecast Data
    daily_forecasts = get_daily_forecast(destination, start_date_obj, num_days)

    # Process Forecast and Generate Context String
    if not daily_forecasts:
        # get_daily_forecast already logged the error
        return f"Could not retrieve a reliable weather forecast for {destination} for {num_days} days starting {start_date_obj.strftime('%b %d')}. Please check the location or try again later."

    full_context = f"Weather & Clothing Tips for {destination} ({num_days} days):\n"
    if len(daily_forecasts) < num_days:
        full_context += f"(Note: Forecast available for only {len(daily_forecasts)} days)\n"


    for day_forecast in daily_forecasts:
        date = day_forecast['date']
        min_t = day_forecast['min_temp']
        max_t = day_forecast['max_temp']
        desc = day_forecast['description']

        # Format date nicely (e.g., Apr 23rd)
        date_obj = date_parser.parse(date)
        date_formatted = date_obj.strftime('%b %d')

        # Format temps, handling N/A
        temp_str = f"High: {max_t}Â°C, Low: {min_t}Â°C"
        if min_t == 'N/A' or max_t == 'N/A':
            temp_str = "Temperature unavailable"


        weather_summary = f"- {date_formatted}: {desc}. {temp_str}."
        clothing_tip = get_clothing_tip_for_range(min_t, max_t, desc)

        full_context += f"{weather_summary}\n  Clothing: {clothing_tip}\n"

    return full_context.strip()

def get_tool_search_results(query: str, tool_name: str) -> str:
    """Helper function to run Google Search and handle results/errors."""
    try:
        results = Google_Search.run(query)
        if results and results != "No good Google Search Result was found":
            # Limit length to avoid overwhelming the LLM context
            return f"{tool_name} context: {results[:600]}..."
        else:
            return f"Couldn't find specific {tool_name} info via search."
    except Exception as e:
        return f"Error searching for {tool_name} information."

def get_cultural_context(destination: str, preferences: str) -> str:
    """Gets cultural context using Google Search."""
    query = f"Unique funny cultural quirks or tips for tourists in {destination} related to {preferences}"
    return get_tool_search_results(query, "Cultural tips")

def get_language_context(destination: str) -> str:
    """Gets language context using Google Search."""
    # Focused query for fun/essential phrases
    query = f"Funny or essential basic local language phrases for tourists in {destination}"
    return get_tool_search_results(query, "Language tips")

def get_safety_context(destination: str) -> str:
    """Gets safety context using Google Search."""
    query = f"Key safety tips for tourists in {destination} (keep it concise)"
    return get_tool_search_results(query, "Safety tips")


def silly_travel_stylist_structured(agent_input: dict) -> Dict:
    """
    Generates a humorous and informative travel briefing using a mix of real-time data and LLM-powered generation.

    This function accepts structured travel input (destination, date, duration, preferences, and budget),
    gathers relevant contextual information (weather forecast, cultural tips, safety, and language),
    and then prompts a Gemini model to generate a `SillyTravelBriefing` response in structured JSON format.

    It uses tool-augmented grounding (Google Search + Weather APIs) and structured output parsing
    with a Pydantic model to ensure valid, helpful, and quirky suggestions for travelers.

    Args:
        agent_input (dict): A dictionary containing:
            - destination (str): The location of the trip.
            - travel_date (str): Start date of the trip.
            - duration (str): Duration string like "5 days", "1 week", etc.
            - preferences (str): Traveler interests and preferences.
            - budget (str, optional): Budget for the trip.

    Returns:
        dict: Either a structured JSON-compatible travel briefing (as a dictionary),
              or an error message if parsing or generation fails.

    Example Output:
        {
            "destination": "Tokyo",
            "weather_tip": "Expect mild spring rain. Pack a cute umbrella!",
            "cultural_tip": "Slurp those noodles loud and proudâ€”itâ€™s polite here.",
            "language_tip": "Learn to say 'Sumimasen'â€”youâ€™ll use it everywhere.",
            "safety_tip": "Tokyo is extremely safe, but donâ€™t jaywalk!",
            "packing_tip": "Bring layersâ€”itâ€™s chilly in the mornings.",
            "activity_recommendation": "Explore themed cafÃ©s and cherry blossoms in Ueno Park."
        }
    """
    try:
        prefs = TravelPreferences(**agent_input)
    except: # Catches Pydantic validation errors too
        return {"error": f"Could not parse input data. Check fields. Error: {e}"}

    if isinstance(prefs.travel_date, str):
        from dateutil import parser
        start_date = parser.parse(prefs.travel_date)
    else:
        start_date = prefs.travel_date

    duration_days = int(re.search(r'\d+', agent_input["duration"]).group())
    end_date = start_date + timedelta(days=duration_days)
    ten_days_from_now = datetime.now() + timedelta(days=10)

    # Gather Context from Tools
    if end_date <= ten_days_from_now:
        weather_context = get_multi_day_weather_and_clothing_context(
            agent_input["destination"],
            start_date,
            agent_input["duration"]
        )
    else:
        prompt = (
        f"I'm planning a trip to {agent_input['destination']} in {agent_input['travel_date']}."
        f"Can you give me tips on what kind of clothes to pack based on the typical weather?"
        )
        weather_context = google_llm.invoke(prompt).content.strip()
    cultural_context = get_cultural_context(agent_input["destination"], agent_input["preferences"])
    language_context = get_language_context(agent_input["destination"])
    safety_context = get_safety_context(agent_input["destination"])

    # Prepare Input for LLM
    combined_context = f"""
    Travel Destination: {agent_input["destination"]}
    Travel Date: {agent_input["travel_date"]}
    Trip Duration: {agent_input["duration"]}
    Traveler Preferences: {agent_input["preferences"]}
    Budget: {agent_input["budget"] or 'Not specified'}

    Gathered Information:
    Weather/Clothing Context: {weather_context}
    Cultural Context: {cultural_context}
    Language Context: {language_context}
    Safety Context: {safety_context}
    """

    # Define Prompt for Structured Output Generation
    prompt_template = ChatPromptTemplate.from_messages([
        ("system",
         "You are the 'Silly Travel Stylist'. Your task is to create a humorous and quirky, yet informative, travel briefing based on the provided context. "
         "Use the gathered information about weather, culture, language, and safety, along with the traveler's preferences. "
         "Inject humor and personality into each field. "
         "Generate the output STRICTLY in the requested JSON format matching the SillyTravelBriefing schema."
         "This JSON must match the `SillyTravelBriefing` structure exactly, including all fields."
         "Focus on being funny but still providing a hint of useful advice within the humor. "
         "Refer to the traveler's preferences when suggesting activities."),
        ("human",
         "Please generate a SillyTravelBriefing JSON for the following travel plan:\n\n{context}")
    ])

    # Create LLM Chain with Structured Output
    # This ensures the LLM output conforms to the Pydantic model
    structured_llm = google_llm.with_structured_output(SillyTravelBriefing)
    chain = prompt_template | structured_llm

    # 5. Invoke Chain and Get Structured Output
    try:
        response = chain.invoke({"context": combined_context})
        return str(response.model_dump())

    except Exception as e:
        # Fallback or error reporting
        error_message = f"Failed to generate the silly travel briefing. The LLM might be having an off day! Error: {e}"
        return {"error": error_message}
````

````python
# Define the structured output schema
class Activity(BaseModel):
    title: str
    description: str
    activity_type: str
    time_of_day: str
    estimated_start_time: str
    estimated_duration: str
    estimated_budget: str

class DailyActivity(BaseModel):
    day: int
    title: str
    daily_tips: str
    activities: List[Activity]

class Itinerary(BaseModel):
    assumptions: List[str]
    general_tips: List[str]
    accomodation: str
    accomodation_budget: str
    transportation: str
    transportation_budget: str
    daily_activities: List[DailyActivity]
    
# Create parser
parser_itinerary = PydanticOutputParser(pydantic_object=Itinerary)

# Should have its own llm instance 
llm_itinerary = ChatGoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=GOOGLE_API_KEY, temperature=1)

# Node as a function that return the task response (detinstiaon information in this example)
def generate_itinerary(agent_input : dict):
    """
    Generates a structured, personalized travel itinerary using Gemini LLM and user-defined trip details.

    The function takes in a dictionary of travel-related inputs such as accommodation, preferences, transportation, and visa info,
    then uses a language model to produce a day-by-day itinerary formatted according to a predefined Pydantic schema (`Itinerary`).

    The output includes realistic, culturally relevant activities and helpful tips tailored to the userâ€™s budget, interests, and location.

    Args:
        agent_input (dict): Must contain the following fields:
            - destination (str): City or region the user is visiting.
            - travel_date (str): Start date of the trip.
            - duration (str): Duration in days.
            - activity_preferences (str): User's activity interests.
            - budget (str): Estimated trip budget.
            - accommodation (str): Lodging information.
            - destination_info (str): Local facts or context.
            - transportation (str): How the user will travel.
            - trip_tips (str): Custom travel advice.
            - visa_info (str): Visa requirement details.

    Returns:
        str: A JSON-formatted itinerary generated by the model and validated using the `Itinerary` Pydantic schema.
              If a required field is missing, an error string is returned.

    Example Output:
        {
            "days": [
                {
                    "date": "April 23, 2025",
                    "activities": ["Visit Emirates Stadium", "Explore Camden Market"],
                    "tips": "Buy tickets for the stadium tour in advance."
                },
                ...
            ],
            "general_tips": "Bring an umbrella and wear water-resistant shoes in London."
        }
    """
    ITINERARY_SYSTEM_PROMPT = """
You are an expert travel assistant that builds custom tourist itineraries.

Your goal is to create a structured itinerary based on:
- accommodation
- the user's travel preferences,
- destination,
- trip duration,
- travel date,
- destination_info,
- transportation,
- trip tips,
- visa information
- budget

Only suggest realistic, enjoyable, and culturally respectful tourist activities. Include food, sightseeing, walking tours, or nature experiences when appropriate.

Respond in the following JSON format:
{format_instructions}
""".strip()

    ITINERARY_HUMAN_PROMPT = """
I'm planning a trip with the following details:

- Accommodation: {accommodation}
- Destination: {destination}
- Travel date: {travel_date}
- Duration: {duration} days
- My preferences: {activity_preferences}
- Budget: {budget}
- Destination Information: {destination_info}
- Transportation: {transportation}
- Travel Tips: {trip_tips}
- Visa information: {visa_info}

Please create a day-by-day itinerary with tips and suggested activities. Make sure to include general tips and any assumptions you made.
""".strip()
    
    required_input_features = ["destination", "travel_date", "duration", "activity_preferences",
                                "budget", "accommodation", "destination_info", "transportation",
                                "trip_tips", "visa_info"]
    
    for feature in required_input_features:
        if feature not in agent_input:
            return f"Agent Input Error: the input must include {feature}."
    
    prompt = ChatPromptTemplate.from_messages([
        ("system", ITINERARY_SYSTEM_PROMPT),
        ("human", ITINERARY_HUMAN_PROMPT),
    ])
    
    formatted_prompt = prompt.format_messages(
        format_instructions=parser_itinerary.get_format_instructions(),
        **agent_input
    )
    response = llm_itinerary.invoke(formatted_prompt)
    result = parser_itinerary.parse(response.content)
    return json.dumps(result.model_dump())
````

## ğŸ§  The Supervisor Agent

This is the brain of the multi-agent architecture. It doesn't generate travel content directly. Instead, it:

- Reads the current state (`TripState`) which includes everything we know about the trip.
- Decides which agent to call next, based on available inputs.
- Sends the right input to that agent.
- Updates the state with new outputs.

Think of it as a project manager coordinating a team of specialists.

### ğŸ§± Why Use `@summarize_context_after_call`

This decorator:
- Wraps agent functions.
- Automatically calls `context_summarizer()` after the agent returns output.
- Ensures new data (like selected transportation or tips) is injected back into the `TripState`.

This helps avoid repeating context parsing logic in each agent.

### ğŸ” Why do we have â€œContext Summarizer that Updates the Shared Stateâ€?
Each time the user types something new or an agent (like visa checker, accommodation recommender, etc.) outputs a result, that new info needs to be injected into the global shared state so all future agents know whatâ€™s already known.
Thatâ€™s exactly what the context_summarizer() does: 
1. Takes old context (TripState) and some new info (e.g., "I want to visit Tokyo on April 20").
2. Sends that to Gemini with a prompt: â€œHereâ€™s the current context. Hereâ€™s the new input. Extract any relevant updates (like destination, travel date, etc.) in JSON.â€
3. Parses the JSON output (e.g., {"destination": "Tokyo", "start_date": "2025-04-20"}).
4. Updates the state in-place:

Now we define teh TripState that we need to pass between agents.

````python
class TripState:
    chat_history: List[str] = None
    user_preferences: str = None
    current_location: str = None
    budget: str = None
    destination: str = None
    visa_info: str = None
    itinerary_draft: str = None
    personalized_itinerary: str = None
    accommodation: str = None
    duration : str = None
    start_date : str = None
    trip_tips : str = None
    destination_activity: str = None
    transportation : str = None
    next_node: str = None # Added for supervisor
    agent_input: str = None # Added for supervisor
    response: str = None
    
    def __init__(self, chat_history: str, **kwargs):
        self.chat_history = [chat_history]
        self.user_preferences = kwargs.get("user_preferences", chat_history)
        self.current_location = kwargs.get("current_location")
        self.budget = kwargs.get("budget")
        self.destination = kwargs.get("destination")
        self.visa_info = kwargs.get("visa_info")
        self.itinerary_draft = kwargs.get("itinerary_draft")
        self.personalized_itinerary = kwargs.get("personalized_itinerary")
        self.accommodation = kwargs.get("accommodation")
        self.duration = kwargs.get("duration")
        self.start_date = kwargs.get("start_date")
        self.trip_tips = kwargs.get("trip_tips")
        self.destination_activity = kwargs.get("destination_activity")
        self.transportation = kwargs.get("transportation")
        self.next_node = kwargs.get("next_node")
        self.agent_input = kwargs.get("agent_input")
        self.response = kwargs.get("response")
        
    def get_chat_history(self):
        return self.chat_history
    
    def __repr__(self):
        return f"""
    chat_history : {self.chat_history}
    user_preferences : {self.user_preferences}
    current_location : {self.current_location}
    budget : {self.budget}
    destination : {self.destination}
    visa_info : {self.visa_info}
    itinerary_draft : {self.itinerary_draft}
    personalized_itinerary : {self.personalized_itinerary}
    accommodation : {self.accommodation}
    duration : {self.duration}
    start_date : {self.start_date}
    trip_tips : {self.trip_tips}
    destination_activity : {self.destination_activity}
    transportation : {self.transportation}
    next_node : {self.next_node}
    agent_input : {self.agent_input}
    response: {self.response}
    """
    
    def __str__(self):
        return f"""
    chat_history : {self.chat_history}
    user_preferences : {self.user_preferences}
    current_location : {self.current_location}
    budget : {self.budget}
    destination : {self.destination}
    visa_info : {self.visa_info}
    itinerary_draft : {self.itinerary_draft}
    personalized_itinerary : {self.personalized_itinerary}
    accommodation : {self.accommodation}
    duration : {self.duration}
    start_date : {self.start_date}
    trip_tips : {self.trip_tips}
    destination_activity : {self.destination_activity}
    transportation : {self.transportation}
    next_node : {self.next_node}
    agent_input : {self.agent_input}
    response: {self.response}
    """
````

````python


#=== Prompts ===

SUPERVISOR_PROMPT = ChatPromptTemplate.from_messages([
("system", """
ğŸ§  SUPERVISOR AGENT PROMPT

ROLE:
You are the Supervisor Agent in a multi-agent travel planning system. You coordinate specialized agents, maintain an evolving shared context, and guide the user through a complete, intelligent trip-planning journey.

You do NOT generate travel content yourself â€” you only orchestrate agent calls and context updates.

ğŸ§° ACCESS:
- `context`: the shared state (see variables below)
- all specialist agents (see formats below)
- `get_user_input`: to collect or confirm information, or deliver results

ğŸ“¦ CONTEXT VARIABLES:
transportation, chat_history, user_preferences, current_location, budget,
destination, visa_info, itinerary_draft, personalized_itinerary,
accommodation, duration, start_date, trip_tips

---

ğŸ“¦ AGENTS & INPUT FORMATS:

get_accommodation:
{{"destination": string, "user preference": string}}

activity_search:
{{"search_query": "a google search query that includes preferences of the user based on destination"}}

destination_recommender:
{{"travel_date": "string in yy-mm-dd format", "duration": string, "budget": string, "accommodation_preferences": string}}

generate_itinerary:
{{
  "destination": string,
  "travel_date": "string in yy-mm-dd format",
  "duration": string,
  "budget": string,
  "accommodation": string,
  "destination_info": string,
  "transportation": string,
  "trip_tips": string,
  "visa_info": string,
  "activity_preferences": "use both user preferences and destination_activities"
}}

silly_travel_stylist_structured:
{{"destination": string, "travel_date": "string in yy-mm-dd format", "duration": string, "preferences": string, "budget": string}}

get_transportation:
{{"origin": string, "destination": string, "transportation_preferences": string, "start_date": string, "duration": string}}

get_location_visa:
{{"origin": string, "destination": string, "other": string}}

get_user_input:
Use this agent to gather missing info, confirm outputs, or deliver the final result.

---

ğŸ¯ PLANNING FLOW:


1. ğŸ“ **Collect Initial Preferences**
   - Ensure these exist in context: `user_preferences`, `current_location`, `budget`, `start_date`, `duration`
    - Otherwise, request them via `get_user_input` with suitable question as string input.

2. ğŸ“ **Recommend Destination (Optional)**
   - If `destination` is missing or user is unsure and `travel_date`, `duration`, `budget` are known and there are preferences about accomodation in user preferences, else ask user to fill them :
     â†’ Call `destination_recommender`.

3. ğŸ“œ **Check Visa Requirements**
   - If `origin` and `destination` are present and `visa_info` is missing:
     â†’ Call `get_location_visa`.

4. ğŸ›ï¸ **Search for Accommodation**
   - Only if `accommodation` is NOT already in context
   - And if `destination` and user accommodation preferences are available:
     â†’ Call `get_accommodation`.

5. ğŸš„ **Plan Transportation**
   - If `origin`, `destination`, `start_date`, and `duration` are present:
     â†’ Call `get_transportation`.

6. ğŸ§­ **Search for Activities**
   - If `duration` and `destination` are present:
     â†’ Call `activity_search`.

7. ğŸ‰ **Generate Fun Summary**
   - Only if `trip_tips` is NOT already in context
   - And if `duration`, `destination`, `travel_date`, `preferences`, and `budget` are present:
     â†’ Call `silly_travel_stylist_structured`.

8. ğŸ§³ **Generate Itinerary**
   - Only if `itinerary_draft` is NOT already in the current context, if it present go to Send Final Output to User section.
   - And if `destination`, `travel_date`, `duration`, `activity_preferences`,
     `budget`, `accommodation`, `destination_info`, `transportation`,
     `trip_tips`, `visa_info` are available:
     â†’ Call `generate_itinerary`.

9. âœ… **Send Final Output to User**
   - If `itinerary_draft` is now available in the context instead of calling agent, do `action` to send the user the output
     â†’ Return it using:
     ```json
     {{"action": "final_response", "response": the context itinerary_draft as json}}
     ```

---

âš ï¸ RULES:

- Only call an agent when all of its required inputs are available in `context`.
- If something is missing, use `get_user_input` to ask for it with reasonable question.
- Never generate travel content â€” delegate all logic to appropriate agents.
- Whenever you want to call an agent, return a JSON with key:
  ```json
  {{"why": "the reason you called this agent very briefly", "agent": "<name_of_the_agent>", "input": {{<agent_input>}} }}

"""),
("human", "Current context: {context}")
])

#=== Supervisor LLM ===

llm_supervisor = ChatGoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=GOOGLE_API_KEY, temperature=0)
llm_context = ChatGoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=GOOGLE_API_KEY, temperature=0)
````

````python
#=== Nodes ===
def supervisor_node(state: TripState):
    """
    The central decision-maker.

    This node invokes the Supervisor Agent (LLM) using the current context
    and determines which specialized agent should be called next.

    If an action like 'final_response' is returned, the graph terminates.
    Otherwise, it outputs the name of the next node and its input.
    """
    context = str(state)
    chain = SUPERVISOR_PROMPT | llm_supervisor
    result = chain.invoke({"context": context})

    try:
        response = extract_json_from_response(result.content)
        if "agent" in response:
            agent_name = response["agent"]
            agent_input = response.get("input")  # Pass user input
            return {"next_node": agent_name, "agent_input": agent_input}
        elif "action" in response:
            return {"next_node": END}
            # return {"next_node": "final_response", "response": response["response"]}
        else:
            raise ValueError(f"Invalid supervisor response: {response}")
    except json.JSONDecodeError:
        print(f"Error decoding JSON.  LLM Response: {result.content}")
        return {"next_node": "get_user_input", "agent_input": "I encountered an error. Please try again."}  # Or some error state
    except ValueError as e:
        print(f"ValueError: {e}")
        return {"next_node": "get_user_input", "agent_input": "I encountered an error processing the request."}


def get_user_input(state: TripState):
    """
    Requests additional information from the user.

    Triggered when required fields are missing in context. 
    Appends the model's question and user's answer to the chat history,
    then calls the context summarizer to update the shared state.
    """
    """Prompts the user for more information."""

    chat_history = state.get_chat_history() #Retrieves the current chat history (a list of prior model â†” user messages) from the TripState.
    chat_history.append(state.agent_input) #Adds the last prompt (typically the LLM's message asking for info) to the chat history.state.agent_input is the message the model just asked the user (e.g., "What is your budget?").
    user_input = input("Model --- " + state.agent_input + "\nUser :\n") #Prompts the human user for input. It prints the model's message (agent_input) and waits for the user to type a reply.
    chat_history.append(user_input) #Stores the user's response into the chat history, so that future nodes and the summarizer can reference it.
    new_context = context_summarizer(state, "\n".join(chat_history[-2:])) #Calls the context_summarizer() with the last model message and the user reply (last 2 turns), so it can extract new info and update TripState.
    new_context.chat_history= chat_history 
    new_context.response= user_input #Saves the updated chat_history and user's reply (response) into the TripState so downstream agents and the supervisor can access the new data.
    return new_context


def context_summarizer(state: TripState, text_input: str):
    # context summarizer
    """
    Updates the TripState based on new user or agent-generated input.

    This node uses an LLM prompt to extract structured context variables
    (e.g. destination, preferences, budget) from unstructured text.
    """
    
    context_str = str(state)
    context_summary_prompt = ChatPromptTemplate.from_messages([
        ("system", """
            You are a context summarizer. Your job is to update the travel planning context using newly received input.
            Look for these keys:
            user_preferences
            current_location
            budget
            destination
            visa_info
            personalized_itinerary
            duration
            start_date (in yy-mm-dd format, if year not provided use 2025)

            Return only those keys that are found or updated. Do NOT return unrelated or unchanged keys.

            Format:
            ```json
            {{"field1": "updated value", "field2": "another value"}}
            ```
        """),
        ("human", """
        Context so far:
        {context}

        New input:
        {text_input}
        """),
    ])
    
    chain = context_summary_prompt | llm_context
    result = chain.invoke({"context": context_str, "text_input": text_input})

    try:
        updates = extract_json_from_response(result.content)
        for key, value in updates.items():
            if hasattr(state, key):
                setattr(state, key, value)
        return state
    except Exception as e:
        print(f"Failed to summarize context: {e}\nRaw output: {result.content}")
        return {"next_node": "get_user_input", "agent_input": "Something went wrong updating the context."}
````

### ğŸ§© Agent Wrappers

The `summarize_context_after_call` decorator is used to **automatically update the system's context** after an agent has returned a result. It wraps agents that modify the `TripState` â€” ensuring their outputs are summarized and properly injected back into the planning flow.

---

### âœ… When to Use  
This wrapper is essential for **agents that contribute new information** to the shared context, such as:\n
- Destination recommendations  
- Visa requirements  
- Accommodation selections  
- Transportation options

> **Why it matters:** It keeps the shared state consistent, up-to-date, and ready for the next agent in the workflow â€” without requiring manual intervention.

---

### âš™ï¸ How It Works  
1. Calls the wrapped agent with user input  
2. Passes the agent's output to the context summarizer  
3. Updates the `TripState` in-place with extracted values  
4. Returns the updated state for the supervisor to assess next steps

---

### In Summary  
The wrapper enables **seamless context propagation** in a modular, multi-agent system â€” ensuring each step in the itinerary-building process builds meaningfully on the last.


````python
# === Agent Wrappers ===

def get_next_node(state: TripState):
    return state.next_node


def summarize_context_after_call(agent_func):
    def wrapper(state: TripState):
        # print(f"â³ Calling wrapped agent: {agent_func.__name__}")
        
        # Call the original agent with state.agent_input
        output = agent_func(state.agent_input)

        # Update the context based on agent output
        context_summarizer(state, json.dumps(output))  # Update in-place

        # Optionally store response in state
        state.response = output

        return state
    return wrapper

@summarize_context_after_call
def agent_destination_wrapper(input_data):
    """
    Calls the destination recommender agent.

    Based on user travel date, budget, and preferences, it suggests
    suitable destination cities. Context is updated with destination data.
    """
    return destination_recommender(input_data)

@summarize_context_after_call
def get_location_visa_wrapper(input_data):
    """
    Calls the visa agent.

    It provides visa eligibility info based on the origin and destination.
    Returns tips about eVisa, visa-on-arrival, or exemption.
    Context is updated with visa status.
    """
    return get_location_visa(input_data)

@summarize_context_after_call
def acc_agent_wrapper(input_data):
    """
    Calls the accommodation agent to retrieve hotel suggestions
    based on user destination and lodging preferences.
    Automatically updates context after response.
    """
    acc_agent = AccommodationSearchAgent()
    return acc_agent(input_data)

def silly_travel_stylist_structured_wrapper(state):
    """
    Calls the travel stylist agent.

    It creates a fun, quirky travel summary using weather forecasts,
    cultural facts, safety tips, and language phrases for the destination.
    Updates context with `trip_tips`.
    """
    return {"trip_tips": silly_travel_stylist_structured(state.agent_input)}

def generate_itinerary_wrapper(state):
    """
    Calls the itinerary generator agent.

    It uses all available context (destination, accommodation, transport,
    activities, budget, tips, etc.) to generate a structured, day-by-day
    travel itinerary. The itinerary is saved in the context.
    """
    result = generate_itinerary(state.agent_input)
    state.itinerary_draft = result
    return {"itinerary_draft": result}

@summarize_context_after_call
def get_transportation_wrapper(input_data):
    """
    Calls the transportation agent.

    It recommends transportation modes (e.g., flight, train, bus) and
    includes cost estimations and descriptions. Context is updated accordingly.
    """
    return get_transportation(input_data)

def activity_search_wrapper(state:TripState):
    """
    Calls the activity search agent.

    It performs a Google search using user preferences and destination,
    then returns top activities or experiences as LLM summaries.
    Updates context with `destination_activity`.
    """
    return {"destination_activity": activity_search(state.agent_input)}

def start_node(state:TripState):
    """
    The entry point of the graph.

    This node takes the initial user message from the chat history,
    summarizes its contents, and initializes the context (TripState)
    with any extractable fields such as destination, duration, preferences, etc.
    """
    state = context_summarizer(state, state.chat_history[0])
    return state
````

# **Architecture Overview**

### ğŸ§© LangGraph Architecture

LangGraph is used to orchestrate the execution of agent nodes. It connects nodes (agents) and defines flow logic between them.
- `StateGraph(TripState)`: creates a graph that maintains evolving context. StateGraph keeps the shared TripState and manages transitions.
- `add_node(name, fn)`: adds an agent node. Each Node is an LLM-based or tool-based function like get_accommodation, get_transportation, etc.
- `add_conditional_edges(...)`: defines how the supervisor decides the next node.
- `set_entry_point('start')`: sets the starting node for execution.

````python

# Create the graph
graph = StateGraph(TripState)

graph.add_node("start", start_node)  # Start goes to supervisor
graph.add_node("supervisor", supervisor_node)  # Add the supervisor node
graph.add_node("get_user_input", get_user_input)
graph.add_node("activity_search", activity_search_wrapper)
graph.add_node("destination_recommender", agent_destination_wrapper)
graph.add_node("get_location_visa", get_location_visa_wrapper)
graph.add_node("get_accommodation", acc_agent_wrapper)
graph.add_node("silly_travel_stylist_structured", silly_travel_stylist_structured_wrapper)
graph.add_node("generate_itinerary", generate_itinerary_wrapper)
graph.add_node("get_transportation", get_transportation_wrapper)

# Define Flow
graph.add_edge("start", "supervisor")
graph.add_conditional_edges("supervisor", get_next_node)

# Route everything back to supervisor
graph.add_edge("get_user_input", "supervisor") # Go back to supervisor after user input
graph.add_edge("activity_search","supervisor")
graph.add_edge("destination_recommender","supervisor")
graph.add_edge("get_location_visa","supervisor")
graph.add_edge("get_transportation","supervisor")
graph.add_edge("get_accommodation","supervisor")
graph.add_edge("silly_travel_stylist_structured","supervisor")
graph.add_edge("generate_itinerary","supervisor")

# Entrypoint of the graph
graph.set_entry_point("start")
# Compile the graph
chain = graph.compile()

if SAVE_GRAPH_IMAGE:
    from io import BytesIO
    from PIL import Image
    stream = BytesIO(chain.get_graph().draw_mermaid_png())
    image = Image.open(stream).convert("RGBA")
    image.save("flow_design.png")
````

### Markdown Generator 
Converts a structured itinerary dictionary into a human-readable, well-formatted Markdown travel document.

````python
def generate_markdown(destination : str, itinerary: dict) -> str:
    """
    Converts a structured itinerary dictionary into a human-readable, well-formatted Markdown travel document.

    Args:
        destination (str): The travel destination (e.g., "Bali", "Paris").
        itinerary (dict): A dictionary containing the full itinerary, including:
            - assumptions (List[str])
            - general_tips (List[str])
            - transportation (str)
            - transportation_budget (str)
            - accomodation (str)
            - accomodation_budget (str)
            - daily_activities (List[dict]) with:
                - day (int)
                - title (str)
                - daily_tips (str)
                - activities (List[dict]) with:
                    - title (str)
                    - description (str)
                    - estimated_start_time (str)
                    - estimated_duration (str)
                    - estimated_budget (str)
                    - activity_type (str)

    Returns:
        str: A Markdown-formatted string that presents the full itinerary with emojis, headers, and highlights
             for assumptions, tips, accommodations, and day-by-day activities.
    """
    def bullet_list(items):
        return '\n'.join([f"- {item}" for item in items])

    def format_activity(activity):
        return (
            f"**ğŸ“ {activity['title']}**\n"
            f"> {activity['description']}\n"
            f"> - ğŸ•’ Time: {activity['estimated_start_time']} ({activity['estimated_duration']})\n"
            f"> - ğŸ’¸ Budget: {activity['estimated_budget']}\n"
            f"> - ğŸ”– Type: {activity['activity_type']}\n"
        )

    def format_day(day):
        header = f"## ğŸ“… Day {day['day']}: {day['title']}\n"
        tips = f"**ğŸ’¡ Daily Tips:**\n> {day['daily_tips']}\n"
        activities = "\n".join([format_activity(a) for a in day['activities']])
        return f"{header}\n{tips}\n\n**ğŸ—“ï¸ Activities:**\n\n{activities}\n"

    md = f"# {destination} Travel Itinerary\n\n"

    # Assumptions
    md += "## ğŸ“Œ Assumptions\n"
    md += bullet_list(itinerary.get("assumptions", [])) + "\n\n"

    # General Tips
    md += "## ğŸ§³ General Travel Tips\n"
    md += bullet_list(itinerary.get("general_tips", [])) + "\n\n"

    # Transportation 
    md += "## ğŸ§³ Transportation\n"
    md += itinerary.get("transportation", "Not Provided") + "\n\n"
    md += itinerary.get("transportation_budget", "Not Provided") + "\n\n"
    
    # Accomodation 
    md += "## ğŸ§³ Accomodation\n"
    md += itinerary.get("accomodation", "Not Provided") + "\n\n"
    md += itinerary.get("accomodation_budget", "Not Provided") + "\n\n"

    # Daily Activities
    for day in itinerary.get("daily_activities", []):
        md += format_day(day)

    return md
````
